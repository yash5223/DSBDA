```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=sns.load_dataset('titanic')
```

```python
plt.figure(figsize=(10,8))
plt.title("Distribution of Age vs Gender")
plt.xlabel("Gender")
plt.ylabel("Age")
sns.boxplot(x="sex",y="age",hue="survived" ,data=df)
plt.legend(labels=['NO','YES'])
plt.show()
```
